# Ground living area vs Sales Price

**Chi-square test**

- Null hypothesis: Ground Living Area is independent with Sales Price. In other words, there is no association between Ground Living Area and Sales Price

- Alternative hypothesis: Ground Living Area is not independent with Sales Price. In other words, there is an association between Ground Living Area and Sales Price

```python
var = 'GrLivArea' #Ground Living Area
data = pd.concat([house['SalePrice'], house[var]], axis=1)
data.plot.scatter(x=var, y='SalePrice', ylim=(0,800000))
plt.show()
```

```python
data
```

```python
data.corr()
```

```python
# Chi-square test 
from scipy.stats import chi2_contingency

stat, p, dof, expected = chi2_contingency(data)
 
# interpret p-value
alpha = 0.05
print("p value is " + str(p))
if p <= alpha:
    print('Reject H0: Ground Living Area is not independent with Sales Price of the house')
else:
    print('Accept H0: Ground Living Area is independent with Sales Price of the house')
```
